```python
import numpy as np
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
low = cv2.imread("picXLA.jpg")
plt.imshow(low[:,:,::-1])
```

```python
def adjust_image_gamma(image, gamma=1.0):
    image = np.power(image, gamma)
    max_val = np.max(image.ravel())
    image = image/max_val*255
    image = image.astype(np.uint8)
    return image
```

```python
low_adjusted = adjust_image_gamma(low, 0.45)
plt.imshow(low_adjusted[:,:,::-1])
```

```python
low_adjusted = adjust_image_gamma(low, 0.4)
plt.imshow(low_adjusted[:,:,::-1])
```

```python
low_adjusted = adjust_image_gamma(low, 1.1)
plt.imshow(low_adjusted[:,:,::-1])
```

```python
low_adjusted = adjust_image_gamma(low, 2.0)
plt.imshow(low_adjusted[:,:,::-1])
```
